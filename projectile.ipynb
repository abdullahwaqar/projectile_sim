```python
%matplotlib inline
from matplotlib import pyplot as plt
from projectile import Projectile, init_flight

x0 = 0
y0 = 0
velocity = 10
x10, y10 = init_flight(x0, y0, velocity, 10)
x25, y25 = init_flight(x0, y0, velocity, 25)
x45, y45 = init_flight(x0, y0, velocity, 45)
x30, y30 = init_flight(x0, y0, velocity, 30)
x60, y60 = init_flight(x0, y0, velocity, 60)

# print(x10, y10)
# print(x25, y25)
# print(x45, y45)
# print(x60, y60)

plt.plot(
    x10,
    y10,
    'g-',
    x25,
    y25,
    'go-',
    x45,
    y45,
    'bo-',
    x30,
    y30,
    'ro-',
    x60,
    y60,
    'ko-',
    [0, 12],
    [0, 0],
    'k-'  # ground
)
plt.legend(['25 deg shoot', '10 deg shoot', '45 deg shoot', '30 deg shoot', '60 deg shoot'])
plt.xlabel('X coordinate (m)')
plt.ylabel('Y coordinate (m)')
plt.show()

# Max distance
print(f'Max distance with 60 degree angle:{max(x60):.2f}m')
print(f'Max height with 60 degree angle: {max(y60):.2f}m')
# Max distance
print(f'Max distance with 45 degree angle:{max(x45):.2f}m')
print(f'Max height with 45 degree angle:{max(y45):.2f}m')
```
